```python
import pandas as pd
data={
    'name':['alice','bod','charlie','david','eva'],
    'age':[25,45,30,40,22],
    'department':['sales','engineering','hr','engineering','hod'],
    'salary':[50000,8000,70000,1000,25000]
}
df=pd.DataFrame(data)
df
```

```python
print("first 3 rows:")
df.head(3)
```

```python
print("column names:")
df.columns
```

```python
print("people with salary>50000:")
df[df['salary'] > 50000]
```

```python
print("sorted by age:")
df.sort_values(by='age')
```

```python
grouped=df.groupby('department')['salary'].mean()
print("average salary by department:")
grouped
```

```python
df['salary in laks']=df['salary']/10000
df
```

```python
import matplotlib.pyplot as plt
years=[2019,2020,2021,2022]
revenue=[100,120,130,160]
plt.plot(years,revenue,marker='o',linestyle='-',color='purple')
plt.title("company revenue over years")
plt.xlabel("year")
plt.ylabel("Revenue(in crores)")
plt.xticks(years)
plt.grid(True)
plt.show()
```

```python

 departments = df ['department'].value_counts()
 departments.plot(kind='bar',color='teal')
 plt.title("employee count by department")
 plt.xlabel("department")
 plt.ylabel("count")
 plt.show()
```

```python
ages = df['age']
plt.hist(ages,bins=5,color='coral',edgecolor='black')
plt.title("age distribution")
plt.xlabel("age")
plt.ylabel("Frequency")
plt.show()
```

```python
df.groupby('department').size().plot(kind='pie',autopct='%1.1f%%')
plt.title("department proportion")
plt.ylabel("")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips=sns.load_dataset("tips")
sns.pairplot(tips,hue="sex")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
corr = tips.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
plt.title("boxplot of total bill by day")
plt.show()
```
